<h3>Statistical definitions[In case you need them]</h3> 

<p><b>Mean:</b>In statistics you'll come across slightly different notation than you're probably used to, but the math is exactly the same. The formula to find the sample mean is: = ( Σ xi ) / n. All that formula is saying is add up all of the numbers in your data set ( Σ means “add up” and xi means “all the numbers in the data set).</p>

<p><b>Median:</b>{(n + 1) ÷ 2}th value, where n is the number of values in a set of data. In order to calculate the median, the data must first be ranked (sorted in ascending order). The median is the number in the middle. </p>

<p><b>Mode:</b> The mode of a set of data values is the value that appears most often. If X is a discrete random variable, the mode is the value x (i.e, X = x) at which the probability mass function takes its maximum value.</p>



```
import statistics as stats

print(stats.mean([2,5,6,9]))
print(stats.median(([1,2,3,8,9])))
print(stats.mode([2,5,3,2,8,3,9,4,2,5,6]))
```
