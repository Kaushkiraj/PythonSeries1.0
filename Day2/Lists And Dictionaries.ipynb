### How to access elements from a list?

There are various ways in which we can access the elements of a list.

#### List index

```python
from IPython.display import Image
Image('images/index list.jpg')
```

We can use the index operator [] to access an item in a list. In Python, indices start at 0. So, a list having 5 elements will have an index from 0 to 4.

Trying to access indexes other than these will raise an IndexError. The index must be an integer. We can't use float or other types, this will result in TypeError.

```python
# List indexing

my_list = ['p', 'r', 'o', 'b', 'e']

# Output: p
print(my_list[0])

# Output: o
print(my_list[2])

# Output: e
print(my_list[4])
```

```python
from IPython.display import Image
Image('images/nested list.jpg')
```

Nested lists are accessed using nested indexing.

```python
# Nested List
n_list = ["Happy", [2, 0, 1, 5]]

# Nested indexing
print(n_list[0][1])

print(n_list[1][3])
```

```python
# Error! Only integer can be used for indexing
print(my_list[4.0])
```

#### Negative indexing

```python
# Negative indexing in lists
my_list = ['p','r','o','b','e']

print(my_list[-1])

print(my_list[-5])
```

### How to slice lists in Python?

We can access a range of items in a list by using the slicing operator :(colon).

```python
# List slicing in Python

my_list = ['p','r','o','g','r','a','m','i','z']

# elements 3rd to 5th
print(my_list[2:5])

# elements beginning to 4th
print(my_list[:-5])

# elements 6th to end
print(my_list[5:])

# elements beginning to end
print(my_list[:])
```

### How to change or add elements to a list?

```python
# Correcting mistake values in a list
odd = [2, 4, 6, 8]

# change the 1st item    
odd[0] = 1            

print(odd)

# change 2nd to 4th items
odd[1:4] = [3, 5, 7]  

print(odd)                   
```

We can add one item to a list using the append() method or add several items using extend() method.

```python
from IPython.display import Image
Image('images/list append.png')
```

```python
# Appending and Extending lists in Python
odd = [1, 3, 5]

odd.append(7)

print(odd)
```

```python
odd.extend([9, 11, 13])

print(odd)
```

We can also use + operator to combine two lists. This is also called concatenation.

The * operator repeats a list for the given number of times

```python
# Concatenating and repeating lists
odd = [1, 3, 5]

print(odd + [9, 7, 5])
```

```python
print(["re"] * 3)
```

Furthermore, we can insert one item at a desired location by using the method insert() or insert multiple items by squeezing it into an empty slice of a list.

```python
# Demonstration of list insert() method
odd = [1, 9]
odd.insert(1,3)

print(odd)
```

```python
odd[2:2] = [5, 7]

print(odd)
```

### How to delete or remove elements from a list?

```python
# Deleting list items
my_list = ['p', 'r', 'o', 'b', 'l', 'e', 'm']

# delete one item
del my_list[2]

print(my_list)
```

```python
# delete multiple items
del my_list[1:5]

print(my_list)
```

```python
# delete entire list
del my_list
```

```python
# Error: List not defined
print(my_list)
```

We can use remove() method to remove the given item or pop() method to remove an item at the given index.

```python
my_list = ['p','r','o','b','l','e','m']
my_list.remove('p')

print(my_list)
```

The pop() method removes and returns the last item if the index is not provided. This helps us implement lists as stacks (first in, last out data structure).

```python
print(my_list.pop(1))
```

```python
print(my_list)
```

```python
print(my_list.pop())
```

```python
print(my_list)
```

We can also use the clear() method to empty a list.

```python
my_list.clear()

print(my_list)
```

### Python List Methods

Methods that are available with list objects in Python programming are tabulated below.

They are accessed as list.method(). Some of the methods have already been used above.


Python List Methods
append() - Add an element to the end of the list
extend() - Add all elements of a list to the another list
insert() - Insert an item at the defined index
remove() - Removes an item from the list
pop() - Removes and returns an element at the given index
clear() - Removes all items from the list
index() - Returns the index of the first matched item
count() - Returns the count of the number of items passed as an argument
sort() - Sort items in a list in ascending order
reverse() - Reverse the order of items in the list
copy() - Returns a shallow copy of the list

```python
# Python list methods
my_list = [3, 8, 1, 6, 0, 8, 4]
print(my_list)
```

index() - Returns the index of the first matched item

```python
print(my_list.index(8))
```

count() - Returns the count of the number of items passed as an argument

```python
print(my_list.count(8))
```

sort() - Sort items in a list in ascending order

```python
my_list.sort()
print(my_list)
```

reverse() - Reverse the order of items in the list

```python
my_list.reverse()
print(my_list)
```

## DICTIONARY

### How to create a dictionary?

```python
#initialisation

my_dict = {}
print(type(my_dict))
```

You can create a Python dictionary by enclosing a comma-separated list of key-value pairs in curly braces { }. A colon “:” separates each key from its associated value:
d = {
    <key>: <value>,
    <key>: <value>,
      .
      .
      .
    <key>: <value>
}



```python
# dictionary with integer keys

my_dict = {1:'apple', 2:'ball'}
print(my_dict)
```

```python
# dictionary with mixed keys

my_dict = {'name':'John', 1:[2,4,3] }
print(my_dict)
```

We can also use the built-in function dict( ) to create a dictionary, like so:

```python
# using dict()

my_dict = dict({1:'apple', 2:'ball'})
print(my_dict)
```

### How to access elements from a dictionary?

```python
from IPython.display import Image
Image('images/access value.png')
```

```python
dict = {'Name': 'Zara', 'Age': 7, 'Class': 'First'}

print ("dict['Name']:", dict['Name'])
```

```python

print ("dict['Age']:", dict.get('Age'))
```

### How to add or change elements in a dictionary?

Since Python dictionaries are mutable, you can add elements or even change the values of the existing items contained in a dictionary.

For example, dict[key] = ‘value’. To update an existing value in a dictionary, you have to use the built-in update( ) method. You must remember that while adding a value to a dictionary if the value already exists, the value gets updated otherwise a new key with the value is added to the dictionary.

```python
from IPython.display import Image
Image('images/add key-value.png')
```

```python
my_dict = {'name':'Jack', 'age': 26}
print(my_dict)
```

```python
# update value

my_dict['age'] = 27
print(my_dict)
```

```python
# add item

my_dict['address'] = 'Downtown'
print(my_dict)
```
